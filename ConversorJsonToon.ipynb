```python

import json
from toon import encode, decode
import time

# 1. Estructura de datos JSON de ejemplo (Lista uniforme de objetos)
json_data = {
    "productos": [
        {"id": 101, "nombre": "Cafetera Express", "precio": 120.50, "stock": 5},
        {"id": 102, "nombre": "Molino Manual", "precio": 35.99, "stock": 12},
        {"id": 103, "nombre": "Taza de Cerámica", "precio": 15.00, "stock": 50}
    ],
    "pedido_id": "ORD-2025-456",
    "cliente": {
        "nombre": "Elena G.",
        "pais": "Chile"
    }
}

# 2. Conversión a TOON
toon_output = encode(json_data)

# 3. Mostrar resultados y longitudes
print("--- DATOS ORIGINALES (JSON) ---")
# Usamos time.sleep para que la salida en Colab sea más legible después de la instalación.
time.sleep(1)
json_str = json.dumps(json_data)
print(json.dumps(json_data, indent=2))
print(f"Longitud de la cadena JSON (caracteres): {len(json_str)}")
print("-" * 50)


print("\n--- DATOS CONVERTIDOS (TOON) ---")
print(toon_output)
print(f"Longitud de la cadena TOON (caracteres): {len(toon_output)}")
print("-" * 50)
```

```python

from toon import encode, decode
import json

# Asumiendo que 'toon_output' se generó en la celda anterior
# Si no lo hiciste, pega la línea 'toon_output = encode(json_data)' aquí.

# Decodificación TOON a la estructura original
decoded_data = decode(toon_output)

print("--- DATOS DECODIFICADOS (JSON) ---")
print(json.dumps(decoded_data, indent=2))

# Comprobación de que la estructura es idéntica
# Es importante redefinir json_data si la celda 2 no se ejecutó previamente en esta sesión.
json_data_check = {
    "productos": [
        {"id": 101, "nombre": "Cafetera Express", "precio": 120.50, "stock": 5},
        {"id": 102, "nombre": "Molino Manual", "precio": 35.99, "stock": 12},
        {"id": 103, "nombre": "Taza de Cerámica", "precio": 15.00, "stock": 50}
    ],
    "pedido_id": "ORD-2025-456",
    "cliente": {
        "nombre": "Elena G.",
        "pais": "Chile"
    }
}

print(f"\n¿La estructura decodificada es igual a la original? {json_data_check == decoded_data}")
```
