```python

# Celda 2: Carga Dinámica de TODOS los GeoJSONs

import os
import geopandas as gpd
import pandas as pd # <-- ¡Necesario para la concatenación!

# La carpeta de trabajo que ya definimos
DRIVE_FOLDER = '/content/drive/MyDrive/Proyectos_Geo'

# Diccionario para almacenar TODAS las capas GeoJSON
all_merida_layers = {}

# 1. Listar todos los archivos GeoJSON en la carpeta
# Esto se asegura de capturar los 36 archivos que me listaste
archivos_encontrados = [f for f in os.listdir(DRIVE_FOLDER) if f.endswith('.geojson')]

print(f"Iniciando carga de {len(archivos_encontrados)} vectores...")

# 2. Bucle para cargar CADA archivo encontrado
for filename in archivos_encontrados:
    file_path = os.path.join(DRIVE_FOLDER, filename)

    # Usamos el nombre sin la extensión (.geojson) como clave
    layer_name = filename.replace('.geojson', '')

    try:
        all_merida_layers[layer_name] = gpd.read_file(file_path)
    except Exception as e:
        print(f"   ❌ ERROR al cargar {filename}. Error: {e}")

# 3. Separar las capas principales y crear el GeoDataFrame de límites internos
estado_merida = all_merida_layers.get('merida_osm_vector')
municipio_libertador = all_merida_layers.get('merida_libertador_osm_vector')

# Creamos una lista de GeoDataFrames que NO son el estado ni el municipio Libertador (los límites internos)
# Esto incluye a todos los demás municipios y parroquias que cargaste.
parish_and_municipio_layers = [
    gdf for name, gdf in all_merida_layers.items()
    if name not in ['merida_osm_vector', 'merida_libertador_osm_vector']
]

# Unimos todos los límites internos en un solo GeoDataFrame (para dibujar las líneas)
if parish_and_municipio_layers and estado_merida is not None:
    # EL pd.concat AHORA FUNCIONARÁ.
    all_boundaries = gpd.GeoDataFrame(pd.concat(parish_and_municipio_layers, ignore_index=True), crs=estado_merida.crs)
    print(f"\n✅ Total de {len(all_merida_layers)} vectores cargados en memoria.")
else:
    all_boundaries = None
    print("\n⚠️ Advertencia: Solo se cargaron el estado y/o municipio, no hay límites internos para concatenar.")
```

```python

# Celda 4: Carga y Unión de Datos Temáticos (CORREGIDA)

import pandas as pd
import numpy as np

# 1. Crear un DataFrame de datos temáticos de ejemplo
# Mantenemos 'layer_name' en este DataFrame para la unión
data_tematica = {
    'layer_name': [
        'merida_AlbertoAdriani_osm_vector',
        'merida_AndresBello_osm_vector',
        'merida_Libertador_OsunaRodriguez_osm_vector',
        'merida_Libertador_JuanRodriguezSuarez_osm_vector',
        'merida_libertador_osm_vector',
        'merida_osm_vector'
    ],
    'poblacion_estimada_2024': [
        80000, 15000, 12000, 18000,
        250000,
        900000
    ],
    'densidad_km2': [
        np.random.randint(50, 200), np.random.randint(10, 50),
        np.random.randint(100, 500), np.random.randint(100, 600),
        np.random.randint(200, 1000),
        np.random.randint(50, 150)
    ]
}
df_tematico = pd.DataFrame(data_tematica)

# 2. Unir los datos temáticos a los GeoDataFrames
print("Iniciando unión de datos temáticos a las capas Geoespaciales...")

# Iteramos sobre el diccionario de todas las capas cargadas
for layer_name, gdf in all_merida_layers.items():

    # 2a. Añadir la columna de clave al GeoDataFrame si no existe
    # Esto asegura que ambos DataFrames tengan la columna 'layer_name'
    if 'layer_name' not in gdf.columns:
        gdf['layer_name'] = layer_name

    # 2b. Buscar los datos temáticos asociados a esta capa
    # Filtramos el DataFrame temático para obtener SÓLO los datos que coinciden con el nombre de la capa.
    fila_tematica = df_tematico[df_tematico['layer_name'] == layer_name]

    if not fila_tematica.empty:
        # 2c. Realizar la unión (CORRECCIÓN CLAVE: right_on='layer_name')
        # Unimos el GDF con el DF temático usando 'layer_name' en ambos lados.
        all_merida_layers[layer_name] = gdf.merge(
            fila_tematica,
            left_on='layer_name', # Columna del GeoDataFrame
            right_on='layer_name', # Columna del DataFrame temático (ambos son tipo object/str)
            how='left'
        )
        print(f"   -> Datos temáticos unidos a '{layer_name}'.")
    else:
        print(f"   -> No se encontraron datos temáticos para '{layer_name}'.")

# Opcional: Actualizar las variables principales
estado_merida = all_merida_layers.get('merida_osm_vector')
municipio_libertador = all_merida_layers.get('merida_libertador_osm_vector')

print("\n✅ Unión de datos temáticos completada. Lista para la Celda 5.")
```

```python

# Celda 4: Carga y Unión de Datos Temáticos para TODOS los 36 Vectores

import pandas as pd
import numpy as np
import geopandas as gpd

# 1. Obtener la lista COMPLETA de nombres de capa
# Excluimos los nombres que no son límites internos (estado y municipio principal)
all_layer_names = [
    name for name in all_merida_layers.keys()
    if name not in ['merida_osm_vector', 'merida_libertador_osm_vector']
]

# 2. Simular datos para TODOS los límites internos
# Generamos población entre 1,000 y 50,000 (para municipios)
# y densidad entre 50 y 800 (para zonas más pobladas)
np.random.seed(42) # Fija la semilla para resultados repetibles

data_tematica = {
    'layer_name': all_layer_names,
    'poblacion_estimada_2024': np.random.randint(1000, 50000, size=len(all_layer_names)),
    'densidad_km2': np.random.randint(50, 800, size=len(all_layer_names))
}
df_tematico = pd.DataFrame(data_tematica)

# **Añadimos datos para el Municipio Libertador y el Estado Mérida (opcional, para ser exhaustivos)**
df_tematico = pd.concat([df_tematico, pd.DataFrame({
    'layer_name': ['merida_libertador_osm_vector', 'merida_osm_vector'],
    'poblacion_estimada_2024': [250000, 900000],
    'densidad_km2': [350, 100]
})], ignore_index=True)


# 3. Unir los datos temáticos a los GeoDataFrames
print("Iniciando unión de datos temáticos a las 36 capas Geoespaciales...")

for layer_name, gdf in all_merida_layers.items():

    # 3a. Aseguramos que el GDF tenga la columna 'layer_name' para la unión
    if 'layer_name' not in gdf.columns:
        gdf['layer_name'] = layer_name

    # 3b. Realizar la unión con el DataFrame temático completo
    # Usamos la columna 'layer_name' en ambos DataFrames (solución al error previo)

    # Usamos merge para añadir las columnas de población
    all_merida_layers[layer_name] = gdf.merge(
        df_tematico,
        left_on='layer_name',
        right_on='layer_name',
        how='left' # 'left' asegura que la geometría se mantiene incluso si no hay datos
    )
    print(f"   -> Datos temáticos unidos a '{layer_name}'.")

# 4. Actualizar las variables principales
estado_merida = all_merida_layers.get('merida_osm_vector')
municipio_libertador = all_merida_layers.get('merida_libertador_osm_vector')

print("\n✅ Unión de datos temáticos para todas las capas completada. Lista para la Celda 5.")
```

```python

# Celda 6: Instalación y Mapa Interactivo con Folium

# 1. Instalación de Folium (si no está instalado)
!pip install folium

import folium
import pandas as pd
import geopandas as gpd
from IPython.display import display # Importante para mostrar el mapa en Jupyter/Colab

# 2. Preparación de los datos para el GeoDataFrame interactivo
# Excluimos el límite del estado para el GeoJson principal, pero lo usaremos para centrar.
gdfs_a_dibujar = [
    gdf for name, gdf in all_merida_layers.items()
    if name != 'merida_osm_vector' # Excluir el límite del estado
]

if not gdfs_a_dibujar:
    print("❌ ERROR: No se encontraron GeoDataFrames internos para crear el mapa interactivo.")
else:
    # Concatenar todos los GeoDataFrames internos (que ahora tienen datos de población)
    interactive_gdf = gpd.GeoDataFrame(pd.concat(gdfs_a_dibujar, ignore_index=True), crs=gdfs_a_dibujar[0].crs)

    # 3. Determinar el centro del mapa
    # Usaremos el GeoDataFrame completo de los límites internos para centrar la vista
    center_lat = interactive_gdf.geometry.centroid.y.mean()
    center_lon = interactive_gdf.geometry.centroid.x.mean()

    # 4. Crear el mapa base (centrado en Mérida)
    # Usamos CartoDB Positron como fondo, que es limpio y profesional.
    m = folium.Map(location=[center_lat, center_lon], zoom_start=9, tiles="cartodbpositron")

    # 5. Añadir las capas interactivas (GeoJson)
    folium.GeoJson(
        interactive_gdf,
        name='Municipios y Parroquias Temáticas',
        # El estilo lo hacemos simple; el color se puede mapear como en matplotlib, pero lo dejaremos fijo por ahora.
        style_function=lambda x: {
            'fillColor': '#FFA500', # Color base de relleno (Naranja)
            'color': 'black',       # Color del borde
            'weight': 0.8,
            'fillOpacity': 0.5
        },
        # Añadir la funcionalidad de Tooltip (ventana emergente al pasar el mouse)
        tooltip=folium.GeoJsonTooltip(
            fields=['layer_name', 'poblacion_estimada_2024', 'densidad_km2'],
            aliases=['Capa:', 'Población Estimada:', 'Densidad (km²):'],
            localize=True,
            sticky=False,
            labels=True,
            style="""
                background-color: #F0EFEF;
                color: #585858;
                font-family: monospace;
                font-size: 10pt;
                padding: 5px;
            """
        )
    ).add_to(m)

    # 6. Mostrar el mapa
    print("Generando mapa interactivo...")
    display(m)
    print("\n✅ Mapa interactivo generado. Haz zoom y pasa el mouse sobre los polígonos para ver los datos simulados.")
```
